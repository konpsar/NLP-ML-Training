```python
import pandas as pd

data = pd.read_json("./x1.json")
labels = list(data['is_sarcastic'])
print(data)
```

```python
headlines = list(data['headline'])
labels = list(data['is_sarcastic'])
print(headlines,
     labels)
```

### Import Libraries

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

### Tokenize the data & Pad Sequences

```python
tokenizer = Tokenizer(oov_token = "<oov>")
tokenizer.fit_on_texts(headlines)
```

```python
word_index = tokenizer.word_index
print(word_index)
```

```python
seqs = tokenizer.texts_to_sequences(headlines) # create seqs of headlines
padded_seqs = pad_sequences(seqs, padding="post") # post-pad seqs
print(padded_seqs)
print(padded_seqs[0])

```
